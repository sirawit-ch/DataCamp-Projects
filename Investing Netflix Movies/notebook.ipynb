```python
# Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt

# Read in the Netflix CSV as a DataFrame
netflix_df = pd.read_csv("netflix_data.csv")
```

```python
# What was the most frequent movie duration in the 1990s? Save an approximate answer as an integer called duration (use 1990 as the decade's start year).
## explore data
netflix_df.head()
```

```python
# filter the data for movies released in the 1990s
netflix_1990s_df = netflix_df[(netflix_df["release_year"] >= 1990) & (netflix_df["release_year"] < 2000)]
```

```python
netflix_1990s_df.head()
```

```python
# plot for the most frequent movie duration in the 1990s
plt.figure(figsize=(10, 6))
plt.hist(netflix_1990s_df["duration"], bins=20, color='skyblue', edgecolor='black')
plt.title("Distribution of Movie Durations in the 1990s")
plt.xlabel("Duration (minutes)")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

How can I find the most frequent movie duration in netflix_1990s_df ?

```python
# Find the most frequent movie duration in the 1990s using mode()
duration = netflix_1990s_df['duration'].mode()[0]
duration
```

```python
# Count the number of short action movies from the 1990s using .iterrows()
netflix_1990s_action_df = netflix_1990s_df[netflix_1990s_df["genre"] == "Action"]
short_movie_count = 0

for index, row in netflix_1990s_action_df.iterrows():
    if row["duration"] < 90:
        short_movie_count += 1

short_movie_count
```
